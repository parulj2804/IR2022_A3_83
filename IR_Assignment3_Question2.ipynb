```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
wiki_graph = nx.read_edgelist("/content/wiki-vote.txt",  create_using=nx.DiGraph)
```

Find PageRank

```python
pr = nx.pagerank(wiki_graph,max_iter=100)
```

```python
pr
```

Find Hubs and Authority

```python
hubs,authority=nx.hits(wiki_graph,max_iter=100,normalized='True')
hubs,authority
```

Apply Sorting

```python
authority_sorted = {k: v for k, v in sorted(authority.items(), key=lambda item: item[1],reverse=True)}
```

```python
authority_sorted
```

```python
hubs_sorted = {k: v for k, v in sorted(hubs.items(), key=lambda item: item[1],reverse=True)}
```

```python
hubs_sorted
```

```python
pagerank_sorted = {k: v for k, v in sorted(pr.items(), key=lambda item: item[1],reverse=True)}
```

```python
pagerank_sorted
```

Find Top 5 

```python
for k,v in list(hubs_sorted.items())[:5]:
  print(str(k)+":"+str(v))
```

```python
for k,v in list(authority_sorted.items())[:5]:
  print(str(k)+":"+str(v))
```

```python
for k,v in list(pagerank_sorted.items())[:5]:
  print(str(k)+":"+str(v))
```

Plot Graphs

```python
import matplotlib.pyplot as plt
names = list(hubs.keys())[:30]
values = list(hubs.values())[:30]
plt.figure(figsize=(12,7))
plt.scatter(names,values)
fontsize=20
plt.title("Hubs Score", fontdict={'fontsize': fontsize})
plt.xticks(rotation=90)
fontsize1=14
plt.xlabel(" Graph Nodes",fontdict={'fontsize': fontsize1})
plt.ylabel("Score",fontdict={'fontsize': fontsize1})
plt.show()
```
